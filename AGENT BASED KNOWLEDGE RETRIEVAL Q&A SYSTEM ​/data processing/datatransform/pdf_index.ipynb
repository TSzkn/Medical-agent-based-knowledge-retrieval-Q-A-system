```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/ISS/5005/dataset/health_complete.csv')
df2 = pd.read_csv('/content/drive/MyDrive/ISS/5005/dataset/health_2_complete.csv')
df3 = pd.read_csv('/content/drive/MyDrive/ISS/5005/dataset/health_4_complete.csv')
```

```python
df1.drop(df1.columns[[0,1]],axis = 1,inplace = True)
```

```python
df2.drop(df2.columns[0],axis = 1,inplace = True)
```

```python
df3.drop(df3.columns[0],axis = 1,inplace = True)
```

```python
df1['id'] = df1['id'].apply(lambda x : float(f'{int(x)}.0001'))
```

```python
df2['id'] = df2['id'].apply(lambda x : float(f'{int(x)}.0002'))
df3['id'] = df3['id'].apply(lambda x : float(f'{int(x)}.0003'))
```

```python
import ast
def add_prefix(row):
  list = [0.0001]
  row = ast.literal_eval(row)
  for i in row:
    list.append(float(f'{int(i)}.0001'))
  return list
```

```python
df1['parent_data'] = df1['parent_data'].apply(add_prefix)
```

```python
df1
```

```python
def add_prefix_2(row):
  list = [0.0002]
  row = ast.literal_eval(row)
  for i in row:
    list.append(float(f'{int(i)}.0002'))
  return list

df2['parent_data'] = df2['parent_data'].apply(add_prefix_2)
```

```python
def add_prefix_3(row):
  list = [0.0003]
  row = ast.literal_eval(row)
  for i in row:
    list.append(float(f'{int(i)}.0003'))
  return list

df3['parent_data'] = df3['parent_data'].apply(add_prefix_3)
```

```python
df3
```

```python
df = pd.concat([df1,df2,df3])
```

```python
df.to_csv('/content/drive/MyDrive/ISS/5005/dataset/health_all.csv')
```
